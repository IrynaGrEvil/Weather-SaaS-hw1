```python
def get_weather(location, start_date, end_date):
    API_KEY = "R7REWLRV3HEHW4PYKGAFUHMZ2"
    weather_url = f"https://weather.visualcrossing.com/VisualCrossingWebServices/rest/services/timeline/{location}/{start_date}/{end_date}?key={API_KEY}"
    print(weather_url)
    response = requests.get(weather_url)

    if response.status_code != 200:
        return {"error": "Invalid location or API error"}
    
    data = response.json()
    print(data)
    

    return {
        "location": location,
        "temperature": data['currentConditions']['temp'],
        "weather": data['currentConditions']['conditions']
    }
```

```python
df.set_index('datetime', inplace=True)

# Create a figure and axes for plotting
fig, axes = plt.subplots(nrows=4, ncols=2, figsize=(14, 12))

# List of parameters to plot
params = ['temp', 'feelslike', 'humidity', 'pressure', 'windspeed', 'visibility']

# Flatten the axes array to make it easier to iterate
axes = axes.flatten()

# Plot each parameter
for i, param in enumerate(params):
    axes[i].plot(df.index, df[param], marker='o', linestyle='-', label=param)
    axes[i].set_title(f'{param} over time')
    axes[i].set_xlabel('Date')
    axes[i].set_ylabel(param)
    axes[i].grid(True)

# Adjust layout to prevent overlapping labels
plt.tight_layout()

# Show the plot
plt.show()
```
